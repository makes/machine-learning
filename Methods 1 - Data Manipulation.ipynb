Iterating through rows:

```python
for row_id, income in zip(df['Id'], df['Income']):
  if (int(income) < df['Income'].mean()):
    income_level_text = 'below'
  else:
    income_level_text = 'above'
  print("Person %d: %s average income." %(row_id, income_level_text))
```

Loading and filtering a data set, printing out statistics and correlation coefficients.

Data source: https://archive.ics.uci.edu/ml/datasets/chronic_kidney_disease

```python
url = r'http://users.metropolia.fi/~simomake/coursework/ml/kd.csv'
kd = pd.read_csv(url, 
                 sep = ',', 
                 index_col = None,
                 na_values = '?')
affected = kd.where(kd['class'] == 'ckd')
affected.describe()
```

```python
affected.corr()
```

Visualize correlation matrix:

Source: https://machinelearningmastery.com/visualize-machine-learning-data-python-pandas/

```python
import matplotlib.pyplot as plt
names = list(affected.corr())
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, title='Correlation Matrix')
cax = ax.matshow(affected.corr(), vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,11,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```
